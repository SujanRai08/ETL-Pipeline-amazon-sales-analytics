```python
import pandas as pd
```

```python
file_path = 'amazon.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
data.info()
```

```python
data['rating_count'].isnull().sum()
```

```python
print(data.isnull().sum())
```

```python
rating_count_median = data['rating_count'].dropna().str.replace(',','').astype(int).median()
data['rating_count'] = data['rating_count'].fillna(str(rating_count_median))
```

```python
data.dtypes
```

```python
data['rating_count'].unique()
```

```python
data['rating_count'] = data['rating_count'].str.replace(',','')
```

```python
data['rating_count'] = data['rating_count'].astype(float).astype(int)
```

```python
data['discounted_price'] = data['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
data['actual_price'] = data['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
data["discount_percentage"] = data["discount_percentage"].str.replace("%", "").astype(float)
```

```python
data["rating"] = pd.to_numeric(data["rating"], errors="coerce")
```

```python
data['discount_percentage'] = data['discount_percentage'] / 100
```

```python
data.head()
```

```python
category_split = data['category'].str.split('|', expand=True)
```

```python
category_split.columns = ['category_level_1', 'category_level_2', 'category_level_3','category_level_4','category_level_5','category_level_6','category_level_7']
```

```python
data = pd.concat([data, category_split], axis=1)
```

```python
data.tail()
```

```python
data[['product_id', 'category_level_1', 'category_level_2', 'category_level_3','category_level_4','category_level_5', 'category_level_6', 'category_level_7']].head()
```

```python
data[['category_level_1', 'category_level_2', 'category_level_3','category_level_4','category_level_5', 'category_level_6', 'category_level_7']].isna().sum()
```

```python
data['category_level_6'] = data['category_level_6'].fillna('Unknown')
data['category_level_7'] = data['category_level_7'].fillna('Unknown')
```

```python
data['category_level_4'] = data['category_level_4'].fillna(data['category_level_3'].mode()[0])
data['category_level_5'] = data['category_level_5'].fillna(data['category_level_4'].mode()[0])
```

```python
data['category_level_4'] = data['category_level_4'].fillna('Unknown')
data['category_level_3'] = data['category_level_3'].fillna('Unknown')
```

```python
data = data.drop(columns=['category','review_id','review_title','review_content','user_id','user_name'])
```
